```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Prepare the data 
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#ploting data 
plt.plot(x,y)

#show graph
plt.show()
```

```python
x = [0,1,2,5,4]
y = [0,2,3,2,8]
z = [0,3,6,9,12]

#figure size 
plt.figure(figsize=(8,5), dpi= 100) #dot per inch or pixel per inch

#ploting data 
plt.plot(x,y, color ='r', label = 'X and Y')
plt.plot(x,z, 'b--', label = 'X and Z')
plt.plot(x)

#labeling x,y axis
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

#tickmark
plt.xticks([0,2,4,6])

#Provide title 
plt.title("Simple Plot", fontsize = 18)

plt.legend()
plt.show()
```

```python
#prepare data
x = np.linspace(0, 2, 100)
plt.figure(figsize=(8,5), dpi=100)


plt.plot(x, x, label='linear') 
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
 
#Labeling
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
 
# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,0.5,1,1.5,2])
plt.yticks([0,2,4,6,8,10,12,14])
 
#Provide title
plt.title("Erdapy data analysis", fontsize = 15)
 
plt.legend()
plt.show()
```
